## Merge, Join, Concatenate

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df3
```

```python
pd.concat([df1,df2,df3],axis=0)
```

```python
sol = pd.DataFrame({'k1': ['K0','K1','K2','K3'],'k2':['K0','K1','K0','K1'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
sag = pd.DataFrame({'k1': ['K0','K1','K1','K3'],'k2':['K0','K0','K0','K1'],'C':['A0','A1','A2','A3'],'D':['B0','B1','B2','B3']})
```

```python
sol
```

```python
sag
```

```python
pd.merge(sol,sag,on=['k1','k2'],how='inner')
```

```python
pd.merge(sol,sag,on=['k1','k2'],how='outer')
```

```python
sol = pd.DataFrame({'A':['A0','A1','A2'],'B':['B0','B1','B2']},index=[0,1,2])
sag = pd.DataFrame({'C':['C0','C1','C2'],'D':['D0','D1','D2']},index=[0,1,4])
```

```python
sag
```

```python
sol.join(sag)
```

```python
sol.join(sag, how="outer")
```
