## Çoklu değişken grafikleri

```python
fig = plt.figure(figsize=(10,6))

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2+2, label="x**2+2")
ax.plot(x, x**3-1, label="x**3-1")
ax.legend(loc=0)
ax.grid()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=1)
ax.plot(x, x+2, color="red")
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(x, x+1, color="red", linewidth=3)
ax.plot(x, x+2, linewidth=2, linestyle='--')
ax.plot(x, x+3, lw=1, ls='-.', marker='o')
```

```python
fig, ax = plt.subplots(1,3,figsize=(12,4))

ax[0].plot(x, x**2, x, x**3)
ax[0].set_title("default")

ax[1].plot(x, x**2, x, x**3)
ax[1].set_title("x ekseni zoom")
ax[1].set_xlim([2,4])

ax[2].plot(x, x**2, x, x**3)
ax[2].set_title("x ve y ekseni zoom")
ax[2].set_xlim([2,4])
ax[2].set_ylim([4,40])
```
